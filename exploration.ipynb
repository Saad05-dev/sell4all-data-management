```python
print(df['Age'].mean())
print(df['Age'].median())
```

```python
print(df['Customer spendings'].mean())
print(df['Customer spendings'].median())
```

```python
groupByAge = df.groupby('Country').median(numeric_only=True)
print(groupByAge)
```

```python
groupByCutomerSpendingsMean = df.groupby('Country')['Customer spendings'].mean()
groupByCutomerSpendingsSum = df.groupby('Country')['Customer spendings'].sum()
print('Mean for each Country from the Data')
print()
print(groupByCutomerSpendingsMean)
print()
print('Sum for each Country from the Data')
print()
print(groupByCutomerSpendingsSum)
```

```python
groupByCutomerSpendingsMean.plot(kind='bar')
```

```python
groupByCutomerSpendingsSum.plot(kind='bar')
```
